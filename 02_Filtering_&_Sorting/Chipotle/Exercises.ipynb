### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
httml = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(httml, sep='\t')
```

### Step 4. How many products cost more than $10.00?

```python
chipo.item_price = chipo.item_price.str[1:].astype('float')
```

```python
chipo.query('item_price > 10')['item_price'].count()
#or
len(chipo[chipo['item_price'] > 10])
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo.pivot_table('item_price', index='item_name').query('item_price > 10')
# wrong approach - ot calculates mean
```

```python
chipo.pivot_table('quantity', index=['item_name', 'item_price']).query('quantity == 1')
```

```python
# Check why one item name has multiple prices:
chipo[chipo.item_name == 'Barbacoa Bowl']
# it seams that one item can have mutliple choice_descriptions and prices as result
```

```python
# Solution provided by exercise
chipo.drop_duplicates(['item_name', 'quantity']).query('quantity == 1')[['item_name', 'item_price']].sort_values('item_price', ascending=False)
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values('item_name')
chipo.drop_duplicates(['item_name', 'quantity']).query('quantity == 1')[['item_name', 'item_price']].sort_values('item_name', ascending=False)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
list_of_items = chipo.drop_duplicates(['item_name', 'quantity']).query('quantity == 1')[['item_name', 'item_price']]
most_exp_item = list_of_items.sort_values('item_price', ascending=False)[['item_name']].iloc[0][0]
chipo.query('item_name == @most_exp_item').quantity.sum()
```

```python
# different approach. Find the most expensive item and check it's quantity
chipo.sort_values('item_price',ascending=False).head(1)
```

### Step 8. How many times were a Veggie Salad Bowl ordered?

```python
chipo.query('item_name == "Veggie Salad Bowl"')['order_id'].count()
```

### Step 9. How many times people orderd more than one Canned Soda?

```python
chipo.query('item_name== "Canned Soda" & quantity > 1').order_id.count()
# same but slower
len(chipo[(chipo.item_name == "Canned Soda") & (chipo.quantity > 1)])
```
