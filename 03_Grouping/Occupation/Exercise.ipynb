### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
httml = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(httml, sep='|')
users.index = users.user_id
users = users.drop('user_id', axis=1)
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
np.round(users.groupby('occupation').mean())
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
female = users.pivot_table('age', index='occupation', columns='gender', 
                           aggfunc='count', fill_value=0)['F']
male = users.pivot_table('age', index='occupation', columns='gender', 
                         aggfunc='count', fill_value=0)['M']
ratio = (male / (male + female)) * 100
ratio.sort_values(ascending=False)
```

```python
def male_bool(x):
    if x == 'M':
        return 1
    elif x == 'F':
        return 0

users['male'] = users.gender.apply(male_bool)
ratio = (users.groupby('occupation').male.sum() / users.occupation.value_counts()) * 100
ratio.sort_values(ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
min_age = users.groupby('occupation').age.min()
max_age = users.groupby('occupation').age.max()
Total = pd.concat([min_age, max_age], axis=1)
Total.columns = ['Min Age', 'Max Age']
Total
```

```python
users.groupby('occupation').age.aggregate(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.pivot_table('age', index='occupation', columns='gender')
```

### Step 8.  For each occupation present the percentage of women and men

```python
df = users.pivot_table('age', index='occupation', columns='gender', aggfunc='count', fill_value=0)
df['Total'] = df.F + df.M
df.F = (df.F / df.Total) * 100
df.M = (df.M / df.Total) * 100
df.drop(columns='Total', inplace=True)
df
```
